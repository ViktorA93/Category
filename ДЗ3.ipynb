1 Задание

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import math
a = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
a.sort()

# считаем среднее арифметическое
sqrt1 = sum(a)/ len(a)
print('среднее арифметическое для выборки равно:', sqrt1)
# считаем среднее квадратичное отклонение и дисперсии
kv_o = [(x-sqrt1)**2 for x in a] # квадраты отклоенний
#s_kvo = sum(kv_o) / len(a) #среднее_квадратов_отклонений
sm_disp = sum(kv_o) / len(a)
not_sm_disp = sum(kv_o) / (len(a)-1)
print('смещенная дисперсия:', sm_disp, 'несмещенная дисперсия:', not_sm_disp)
s_sqrt_kvo = math.sqrt(sm_disp)
print('среднее квадратичное отклонение:', s_sqrt_kvo)
```

```python
import numpy as np

a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

# Считаем среднее арифметическое
mean = np.mean(a)
print('Среднее арифметическое для выборки равно:', mean)

# Считаем смещённую дисперсию
sm_disp = np.var(a)  # по умолчанию `var` возвращает смещённую дисперсию
print('Смещённая дисперсия:', sm_disp)

# Считаем несмещённую дисперсию
not_sm_disp = np.var(a, ddof=1)  # `ddof=1` делает дисперсию несмещённой
print('Несмещённая дисперсия:', not_sm_disp)

# Считаем среднее квадратическое отклонение
std_dev = np.std(a)  # по умолчанию `std` возвращает стандартное отклонение для смещённой дисперсии
print('Среднее квадратическое отклонение:', std_dev)
```

2 Задание.

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
import math

def comb(n, k):
    return math.comb(n, k)

# Вероятности для первого ящика
total_balls_1 = 8
white_balls_1 = 5
draw_1 = 2

# Вероятности для второго ящика
total_balls_2 = 12
white_balls_2 = 5
draw_2 = 4

# Случай 1: 2 белых из первого ящика и 1 белый из второго
prob_2_white_from_1 = comb(white_balls_1, 2) / comb(total_balls_1, draw_1)
prob_1_white_from_2 = comb(white_balls_2, 1) * comb(total_balls_2 - white_balls_2, draw_2 - 1) / comb(total_balls_2, draw_2)
prob_case_1 = prob_2_white_from_1 * prob_1_white_from_2

# Случай 2: 1 белый из первого ящика и 2 белых из второго
prob_1_white_from_1 = comb(white_balls_1, 1) * comb(total_balls_1 - white_balls_1, draw_1 - 1) / comb(total_balls_1, draw_1)
prob_2_white_from_2 = comb(white_balls_2, 2) * comb(total_balls_2 - white_balls_2, draw_2 - 2) / comb(total_balls_2, draw_2)
prob_case_2 = prob_1_white_from_1 * prob_2_white_from_2

# Итоговая вероятность
total_probability = prob_case_1 + prob_case_2

print(f"Вероятность того, что 3 мяча белые: {total_probability* 100:.2f}%")
```

3 Задание.

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# Вероятности попадания в мишень для каждого спортсмена
p_A1 = 0.9
p_A2 = 0.8
p_A3 = 0.6

# Вероятность того, что выстрел произведен каждым спортсменом (равные шансы)
p_A = 1 / 3

# Общая вероятность попадания в мишень
p_E = (p_A1 * p_A) + (p_A2 * p_A) + (p_A3 * p_A)

# Вероятности того, что выстрел произведен конкретным спортсменом при условии попадания
p_A1_given_E = (p_A1 * p_A) / p_E
p_A2_given_E = (p_A2 * p_A) / p_E
p_A3_given_E = (p_A3 * p_A) / p_E

print(f"Вероятность того, что выстрел произведен первым спортсменом: {p_A1_given_E* 100:.2f}%")
print(f"Вероятность того, что выстрел произведен вторым спортсменом: {p_A2_given_E * 100:.2f}%")
print(f"Вероятность того, что выстрел произведен третьим спортсменом: {p_A3_given_E * 100:.2f}%")
```

4 Задание.

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# Вероятности сдачи сессии по факультетам
p_As = 0.8
p_Bs = 0.7
p_Cs = 0.9

# Вероятность того что студент с определенного факультета
p_A = 1 / 4
p_B = 1 / 4
p_C = 1 / 2

# Общая вероятность сдачи сессии
p_E = (p_As * p_A) + (p_Bs * p_B) + (p_Cs * p_C)

# Вероятности того, что студент сдал сессию по факультетам
p_A_given_E = (p_As * p_A) / p_E
p_B_given_E = (p_Bs * p_B) / p_E
p_C_given_E = (p_Cs * p_C) / p_E

print(f"Вероятность того, что  студент сдавший сессию с А факультета: {p_A_given_E* 100:.2f}%")
print(f"Вероятность того, что  студент сдавший сессию с B факультета: {p_B_given_E * 100:.2f}%")
print(f"Вероятность того, что  студент сдавший сессию с C факультета: {p_C_given_E * 100:.2f}%")
```

5 Задание

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
# Вероятности выхода из строя и их дополнений
p1 = 0.1
p2 = 0.2
p3 = 0.25
q1 = 1 - p1
q2 = 1 - p2
q3 = 1 - p3

# А) Вероятность того, что все три детали выйдут из строя
p_all = p1 * p2 * p3

# Б) Вероятность того, что только две детали выйдут из строя
p_two = (p1 * p2 * q3) + (p1 * q2 * p3) + (q1 * p2 * p3)

# В) Вероятность того, что хотя бы одна деталь выйдет из строя
p_none = q1 * q2 * q3
p_at_least_one = 1 - p_none

# Г) Вероятность того, что от одной до двух деталей выйдут из строя
p_one = (p1 * q2 * q3) + (q1 * p2 * q3) + (q1 * q2 * p3)
p_one_to_two = p_one + p_two

print(f"Вероятность того, что все три детали выйдут из строя: {p_all *100:.2f}%")
print(f"Вероятность того, что только две детали выйдут из строя: {p_two*100:.2f}%")
print(f"Вероятность того, что хотя бы одна деталь выйдет из строя: {p_at_least_one*100:.2f}%")
print(f"Вероятность того, что от одной до двух деталей выйдут из строя: {p_one_to_two*100:.2f}%")
```
